## NYC Taxi Data Ingestion
[Taxi & Limousine Comissionn Trip Record Data](https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page)

```python
dataset_types = ["yellow", "green", "fhv", "fhvhv"]  # List of dataset types to ingest
years = [2023]  # List of years to ingest
months = range(1, 6)  # List of months to ingest - range(start, stop)
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data"

for dataset_type in dataset_types:
    for year in years:
        for month in months:
            month_str = str(month).zfill(2)
            file_name = f"{dataset_type}_tripdata_{year}-{month_str}.parquet"
            source_url = f"{base_url}/{file_name}"
            dest_path = f"/Volumes/etl/raw/nyc_taxi/{dataset_type}/{year}/{file_name}"

            print(f"Processing {file_name}...")
            try:
                dbutils.fs.cp(source_url, dest_path)
            except Exception as e:
                print(f"Error in processing file {file_name}: {e}")
            
            print(f"Processed file {file_name}...")
```
