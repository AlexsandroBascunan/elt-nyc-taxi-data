```python
taxi_type = "yellow"
year = 2023
month = 2 
base_url = "https://d37ci6vzurychx.cloudfront.net/trip-data"

month_str = str(month).zfill(2)
file_name = f"{taxi_type}_tripdata_{year}-{month_str}.parquet"
source_url = f"{base_url}/{file_name}"

dest_path = f"/Volumes/workspace/nyc_taxi/nyc_taxi/{taxi_type}/{year}/{file_name}"

print(f"Processando {file_name}...")
try:
    # Copia o arquivo Parquet remoto para o volume
    dbutils.fs.cp(source_url, dest_path)
    # LÃª o arquivo Parquet do volume com Spark
    df = spark.read.parquet(dest_path)
    df.show()
except Exception as e:
    print(f"Erro ao processar {file_name}: {e}")
```
