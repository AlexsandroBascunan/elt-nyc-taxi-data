```python
from pyspark.sql.functions import *
```

```python
def process_gold_layer_all_taxis_data():
    yellow_silver = spark.table("etl.silver.silver__nyc_taxi_yellow").withColumn("taxi_type", lit("yellow"))
    green_silver = spark.table("etl.silver.silver__nyc_taxi_green").withColumn("taxi_type", lit("green"))

    gold_all_taxis = yellow_silver.unionByName(green_silver)

    gold_all_taxis.write.format("delta") \
        .mode("overwrite") \
        .partitionBy("year", "month") \
        .option("overwriteSchema", "true") \
        .saveAsTable("etl.gold.fact__all_taxi_trips")
```

```python
process_gold_layer_all_taxis_data()
```
