## Business Analysis and Metrics

```python
from pyspark.sql.functions import *
```

### What is the average total amount received in a month considering all the yellow taxis in the fleet?

```python
yellow_taxi_monthly_stats = spark.table("etl.gold.fact__all_taxi_trips").filter(col("taxi_type") == "yellow") \
    .groupBy("year", "month") \
    .agg(
        count("*").alias("total_trips"),
        sum("passenger_count").alias("sum_passengers"),
        round(avg("passenger_count"), 2).alias("avg_passengers"),
        round(sum("total_amount"), 2).alias("sum_revenue"),
        round(avg("total_amount"), 2).alias("avg_total_amount")
    )

display(yellow_taxi_monthly_stats)
```

### What is the average number of passengers (passenger_count) per hour of the day who took a taxi in the month of May, considering all taxis in the fleet? (Considering Yellow and Green Taxi)

```python
avg_passengers_per_hour_on_may = spark.table("etl.gold.fact__all_taxi_trips").filter((col("month") == 5)) \
    .withColumn("hour_of_day", hour(col("pickup_datetime"))) \
    .groupBy("hour_of_day") \
    .agg(
        count("*").alias("total_trips"),
        sum("passenger_count").alias("sum_passengers"),
        round(avg("passenger_count"), 2).alias("avg_passengers"),
        round(sum("total_amount"), 2).alias("sum_revenue"),
        round(avg("total_amount"), 2).alias("avg_total_amount")
    )

display(avg_passengers_per_hour_on_may)
```

### Total Trips by day of the week
Ranges from 1 for a Sunday through to 7 for a Saturday

```python
dayofweek_taxi_stats = spark.table("etl.gold.fact__all_taxi_trips") \
    .withColumn("day_of_week", dayofweek(col("pickup_datetime"))) \
    .groupBy("day_of_week") \
    .agg(
        count("*").alias("total_trips"),
        sum("passenger_count").alias("sum_passengers"),
        round(avg("passenger_count"), 2).alias("avg_passengers"),
        round(sum("total_amount"), 2).alias("sum_revenue"),
        round(avg("total_amount"), 2).alias("avg_total_amount")
    )

display(dayofweek_taxi_stats)
```
